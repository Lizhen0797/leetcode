```python
def ladderLength(beginWord, endWord, wordList):
    """
    :type beginWord: str
    :type endWord: str
    :type wordList: List[str]
    :rtype: int
    """
    def canConvert(s1, s2):
        count = 0
        for i in range(len(s1)):
            if s1[i] != s2[i]:
                count+=1
                if count > 1:
                    return False
        return count == 1
    if endWord not in wordList:
        return 0
    length = len(wordList)
    visited = [False] * length

    idx = -1
    for i in range(len(wordList)):
        if beginWord == wordList[i]:
            idx = i
            break

    if idx != -1:
        visited[idx] = True
    queue = [beginWord]
    count = 0 
    while len(queue) > 0:
        size = len(queue)
        count += 1
        while (size > 0):
            start = queue.pop(0)
            for i in range(len(wordList)):
                if visited[i]:
                    continue
                s = wordList[i]
                if not canConvert(start, s):
                    continue
                if s == endWord:
                    return count + 1
                visited[i] = True
                queue.append(s)
            size -= 1
    return 0
```
