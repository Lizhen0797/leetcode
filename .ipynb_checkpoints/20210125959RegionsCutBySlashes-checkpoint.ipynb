```python
class UnionFind:
    def __init__(self, n):
        self.father = {x:x for x in range(n)}
        #连通分量统计
        self.count = n

    def find(self, x):
        if x != self.father[x]:
            self.father[x] = self.find(self.father[x])
        return self.father[x]

    def merge(self, x, y):
        findx = self.find(x)
        findy = self.find(y)
        if findx != findy:
            self.father[findx] = findy
            self.count -= 1

class Solution:
    def regionsBySlashes(self, grid):
        n = len(grid)
        # 构建并查集 里面共有n * n * 4个联通分量
        uf = UnionFind(n * n * 4)
        for i in range(n):
            for j in range(n):
                # 找到对应的id
                cur = grid[i][j]
                # 将一个网格划分为四个的起点
                start = 4 * (i * n + j)
                # 如果为左斜线
                if cur == "/":
                    # 则 0 和 3为一个联通分量
                    uf.merge(start, start + 3)
                    # 1 2 为一个联通分量
                    uf.merge(start + 1, start + 2)
                # 如果为右斜线
                elif cur == "\\":
                    # 0 1 为联通分量
                    uf.merge(start, start + 1)
                    # 2 3 为联通分量
                    uf.merge(start + 3, start + 2)
                # 如果没有斜线
                elif cur == " ":
                    # 1， 2， 3， 4都为联通分量
                    uf.merge(start, start + 1)
                    uf.merge(start + 1, start + 2)
                    uf.merge(start + 2, start + 3)
                # 对于行来说 1 和 3联通 1 如 1 7 相差6
                if j > 0:
                    uf.merge(start + 3, start - 3)
                # 对于列来说 2 和 0联通 如2 和8 相差4*n+2
                if i > 0:
                    uf.merge(start, start - 4 * n + 2)

        return uf.count
```
