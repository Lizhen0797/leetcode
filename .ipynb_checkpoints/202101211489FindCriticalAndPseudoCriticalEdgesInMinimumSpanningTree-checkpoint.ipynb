```python
def findCriticalAndPseudoCriticalEdges(n, edges):
    """
    :type n: int
    :type edges: List[List[int]]
    :rtype: List[List[int]]
    """
    parent = list(range(n))
    def find(index):
        if index != parent[index]:
            parent[index] = find(parent[index])
        return parent[index]

    def union(index1, index2):
        parent[index2] = index1
    sorted_edges = [[index]+i for index, i in enumerate(edges)]
    sorted_edges = sorted(sorted_edges, key=lambda x:x[-1])
    print(sorted_edges)
    total = 0
    for index, (_,x, y, w) in enumerate(sorted_edges):
        rx, ry = find(x), find(y)
        if rx!=ry:
            union(rx, ry)
            total+=w
    key_edge = []
    no_key_edge = []
    for i, edge in enumerate(sorted_edges):
        (_,cx,cy,cw) = edge
        tmp_edges = sorted_edges[:i] + sorted_edges[i+1:]

        total1 = cw

        parent = list(range(n))
        union(cx, cy)
        for index, (_, x, y, w) in enumerate(tmp_edges):
            rx, ry = find(x), find(y)
            if rx!=ry:
                union(rx, ry)
                total1+=w
        if total != total1:
            continue
        total2 = 0
        parent = list(range(n))
        for index, (_, x, y, w) in enumerate(tmp_edges):
            rx, ry = find(x), find(y)
            if rx!=ry:
                union(rx, ry)
                total2+=w
        if total1 != total2:
            key_edge.append(edge[0])
        else:
            no_key_edge.append(edge[0])
    return [key_edge, no_key_edge]
```
